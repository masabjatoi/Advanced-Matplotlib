```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
plt.style.use("fivethirtyeight")
```

```python
iris = sns.load_dataset("iris")
```

```python
iris["species"].unique()
```

```python
iris["species"].replace({'setosa':0, 'versicolor':1, 'virginica':2} , inplace = True)
```

```python
plt.figure(figsize = (10,6))

plt.scatter(iris["sepal_length"] , iris["petal_length"] , c = iris["species"] , cmap = "plasma" , label = ["setosa:0" , "versicolor:1" , "virginica:2"]  ,alpha = (0.8))
plt.xlabel("sepal length")
plt.ylabel("petal length")
plt.title("Sepal length and petal length mapping ")
plt.grid()

plt.legend(bbox_to_anchor = (1.9,1))
plt.colorbar()
plt.savefig("scatter plot iris.png")
plt.show()
```

```python
batter = pd.read_csv("batter.csv")
```

```python
batter
```

```python
sample= batter.head(100).sample(25 , random_state = 5)
```

```python
plt.figure(figsize= (18,10))

plt.scatter(sample["avg"]  , sample["strike_rate"] , color = "black" , s  = 200 )
plt.axhline(130,color  ="red")
plt.axvline(30,color  ="red")
for i in range(len(sample)):
    plt.text(sample["avg"].values[i],sample["strike_rate"].values[i] , sample["batter"].values[i] , fontdict = {"size":15 , "color":"blue"})
```

```python
fig , ax = plt.subplots(figsize = (10,6))

ax.scatter( sample["avg"] ,sample["strike_rate"] , color = ["red"] , marker = "+")
ax.set_title("Avg and strike rate")
ax.set_xlabel("AVG ")
ax.set_ylabel("Strike rate")
fig.show()
```

```python
fig , ax = plt.subplots(nrows = 2 , ncols = 1 , sharex = True)

ax[0].scatter(batter["avg"] ,batter["strike_rate"] )
ax[1].scatter(batter["avg"] ,batter["runs"] )


ax[0].set_title("Avg vs strike_rate")
ax[0].set_ylabel("Strike_rate")
ax[0].set_xlabel("Avg")

ax[1].set_title("Avg vs  Runs")
```

```python
fig , ax = plt.subplots(nrows = 2 , ncols = 2 ,  figsize = (10,10))

ax[0,0].scatter(batter["avg"] ,batter["strike_rate"] , color = "red")
ax[0,1].scatter(batter["avg"] ,batter["runs"] )
ax[1,0].hist(batter["strike_rate"] )
ax[1,1].hist(batter["avg"]  )

ax[0,0].set_title("Avg vs strike_rate")
ax[0,1].set_title("Avg vs runs")
ax[1,0].set_title("Strike_rate")
ax[1,1].set_title("Avg")

ax[0,0].set_ylabel("strike_rate")
ax[0,0].set_xlabel("Avg ")


ax[1,0].set_ylabel("Strike_rate")

ax[1,1].set_xlabel("Avg")


plt.subplots_adjust(wspace=0.3, hspace=0.4)



plt.show()
```

```python
ipl = pd.read_csv("IPL_Ball_by_Ball_2008_2022.csv")
```

```python
ipl
```

```python
ipl["ballnumber"].unique()
```

```python
temp = ipl[(ipl["ballnumber"].isin([1,2,3,4,5,6])) & (ipl["batsman_run"] == 6)]
```

```python
grid = temp.pivot_table(index = "overs" , columns = "ballnumber" , values = "batsman_run" , aggfunc = "count")
```

```python
plt.figure(figsize = (10,10))
plt.imshow(grid)
plt.yticks(ipl["overs"].unique() , list(range(1,21)))
plt.xticks(np.arange(0,6) , list(range(1,7)))
plt.colorbar()
```
